```python
import pandas as pd
import numpy as np
import os
%matplotlib inline
import matplotlib.pyplot as plt
flu_data = "Resources/Outpatient_Flu_Positive.csv"
flu_df = pd.read_csv(flu_data)
flu_df.head().style.hide_index()
```

```python
#Drop the date_code column
flu2_df = flu_df.drop(columns=['date_code'])
```

```python
#Filter dataframe to only show 2017-2018 data
flu3_df=flu2_df.loc[flu2_df['season'] == '2017-2018']
flu3_df.loc[flu3_df['Respiratory_Virus'] == 'Total_Influenza'].head(100).style.hide_index()
```

```python
#Filter the dataframe based on Total_Influenza and only region not California
flu3_df.loc[flu3_df['Respiratory_Virus'] == 'Total_Influenza'].loc[flu3_df['region'] != 'California'].loc[flu3_df['region'] != 'Upper/Lower Southern'].style.hide_index()
```

```python
flu4_df=flu3_df.groupby(['region']).sum()
```

```python
#Create a Bar Chart that compares the Specimens Tested vs the Number of Positive lab tests by Region
flu3_df=flu3_df.loc[flu3_df['Respiratory_Virus'] == 'Total_Influenza'].loc[flu3_df['region'] != 'California'].loc[flu3_df['region'] != 'Upper/Lower Southern']
flu4_df=flu3_df.groupby(['region']).sum()
#Delete the Percent Positive data from the bar graph
del flu4_df['Percent_Positive']
#Identify Number Positive and Specimens Tested as the column to apply to the bar graph
flu4_df.columns=['Number Positive', 'Specimens Tested']
flu4_df_percent= (flu4_df ['Number Positive'] * 100/ (flu4_df ['Specimens Tested']))
flu4_df_percent= flu4_df_percent.tolist()
print(flu4_df)
#Create the Bar Graph
ax = flu4_df.plot.bar(figsize=(15,10), rot=0)
ax.set_xlabel("Region (2017-2018)")
ax.set_ylabel("Sum of Count")
ax.legend(loc="best")
#Place Percent Positive above the Number Positive bar
rects = ax.patches
# Make some labels.
labels = ["%.0f%%" % (i) for i in flu4_df_percent]
j = 0
for rect, label in zip(rects, labels):
    j=j+1
    if j > len(rects)/2:
        continue
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label,
            ha='center', va='bottom')
#Save Fig
plt.savefig("2017_2018_Regional Flu")
plt.show()
```

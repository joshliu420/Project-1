```python
import requests
import json
from pprint import pprint
import os
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
# open the raw dataset
csv_path = os.path.join("vaccination_raw.csv")
df_raw = pd.read_csv(csv_path, encoding='unicode_escape', header=[0,1,2])
df_raw
```

```python
# make a copy of raw dataset to tailor
df_clean = df_raw.copy()
df_clean
```

```python
# clean the header
df_clean.columns = df_clean.columns.droplevel([0,1])
df_clean
```

```python
# set index by "Names"
df_clean = df_clean.set_index('Names')
```

```python
# delete irrelevant columns
df_clean.drop(["LL","UL","CI","SAMPLE","TARGET"],axis=1, inplace=True)
df_clean
```

```python
# pick up data for year 2014-15
df_2014 = df_clean[["2014-15"]]
df_2014
```

```python
# define each columns 
df_2014.columns = ['< 6 months', '6 months - 17 years', '6 months - 4 years', '5-12 years', 
                   '13-17 years', '³18 years','18-64 years','18-64 years at high risk',
                   '18-64 years not at high risk','18-49 years','18-49 years at high risk',
                   '18-49 years not at high risk','50-64 years','> 65 years','White',
                   'Black','Hispanic','Other Races']
df_2014
```

```python
# delete the overlapped bins 
df_2014_clean = df_2014[['< 6 months','6 months - 4 years','5-12 years','13-17 years',
                        '18-49 years','50-64 years','> 65 years','White','Black','Hispanic','Other Races']]
df_2014_clean = df_2014_clean.reset_index()
df_2014_clean
```

```python
# repeat the steps above to get dataframe for year 2015-16,2016-17,2017-18,2018-19
df_2015 = df_clean[["2015-16"]]
df_2015.columns = ['< 6 months', '6 months - 17 years', '6 months - 4 years', '5-12 years', 
                   '13-17 years', '³18 years','18-64 years','18-64 years at high risk',
                   '18-64 years not at high risk','18-49 years','18-49 years at high risk',
                   '18-49 years not at high risk','50-64 years','> 65 years','White',
                   'Black','Hispanic','Other Races']
df_2015_clean = df_2015[['< 6 months','6 months - 4 years','5-12 years','13-17 years',
                        '18-49 years','50-64 years','> 65 years','White','Black','Hispanic','Other Races']]
df_2015_clean = df_2015_clean.reset_index()
df_2015_clean
```

```python
df_2016 = df_clean[["2016-17"]]
df_2016.columns = ['< 6 months', '6 months - 17 years', '6 months - 4 years', '5-12 years', 
                   '13-17 years', '³18 years','18-64 years','18-64 years at high risk',
                   '18-64 years not at high risk','18-49 years','18-49 years at high risk',
                   '18-49 years not at high risk','50-64 years','> 65 years','White',
                   'Black','Hispanic','Other Races']
df_2016_clean = df_2016[['< 6 months','6 months - 4 years','5-12 years','13-17 years',
                        '18-49 years','50-64 years','> 65 years','White','Black','Hispanic','Other Races']]
df_2016_clean = df_2016_clean.reset_index()
df_2016_clean
```

```python
df_2017 = df_clean[["2017-18"]]
df_2017.columns = ['< 6 months', '6 months - 17 years', '6 months - 4 years', '5-12 years', 
                   '13-17 years', '³18 years','18-64 years','18-64 years at high risk',
                   '18-64 years not at high risk','18-49 years','18-49 years at high risk',
                   '18-49 years not at high risk','50-64 years','> 65 years','White',
                   'Black','Hispanic','Other Races']
df_2017_clean = df_2017[['< 6 months','6 months - 4 years','5-12 years','13-17 years',
                        '18-49 years','50-64 years','> 65 years','White','Black','Hispanic','Other Races']]
df_2017_clean = df_2017_clean.reset_index()
df_2017_clean
```

```python
df_2018 = df_clean[["2018-19"]]
df_2018.columns = ['< 6 months', '6 months - 17 years', '6 months - 4 years', '5-12 years', 
                   '13-17 years', '³18 years','18-64 years','18-64 years at high risk',
                   '18-64 years not at high risk','18-49 years','18-49 years at high risk',
                   '18-49 years not at high risk','50-64 years','> 65 years','White',
                   'Black','Hispanic','Other Races']
df_2018_clean = df_2018[['< 6 months','6 months - 4 years','5-12 years','13-17 years',
                        '18-49 years','50-64 years','> 65 years','White','Black','Hispanic','Other Races']]
df_2018_clean = df_2018_clean.reset_index()
df_2018_clean
```

```python
# output each dataset
df_2014_clean.to_csv("vaccination_2014-15.csv", index=False, header=True)
df_2015_clean.to_csv("vaccination_2015-16.csv", index=False, header=True)
df_2016_clean.to_csv("vaccination_2016-17.csv", index=False, header=True)
df_2017_clean.to_csv("vaccination_2017-18.csv", index=False, header=True)
df_2018_clean.to_csv("vaccination_2018-19.csv", index=False, header=True)
```

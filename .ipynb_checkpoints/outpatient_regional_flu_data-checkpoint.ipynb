```python
import pandas as pd
import numpy as np
import os
%matplotlib inline
import matplotlib.pyplot as plt
flu_data = "Resources/Outpatient_Flu_Outcome_California.csv"
flu_df = pd.read_csv(flu_data)
flu_df.head().style.hide_index()
```

```python
flu2_df = flu_df.drop(columns=['date_code'])
```

```python
flu2_df.rename(columns={"season": "Season", "weekending": "Weekending", "region":"Region"}).head().style.hide_index()
```

```python
flu3_df=flu2_df.loc[flu2_df['season'] == '2017-2018']
flu3_df.loc[flu3_df['Respiratory_Virus'] == 'Total_Influenza'].head(100).style.hide_index()
```

```python
flu3_df["Month"] = flu3_df["weekending"].apply(lambda d: d[:2])
```

```python
flu3_df.head()
```

```python
flu3_df.loc[flu3_df['Respiratory_Virus'] == 'Total_Influenza'].head(100).loc[flu3_df['region'] != 'California'].head(100).style.hide_index()
```

```python
#fig, ax=plt.subplots(figsize=(15, 10));
flu3_df=flu3_df.loc[flu3_df['Respiratory_Virus'] == 'Total_Influenza'].head(100).loc[flu3_df['region'] != 'California'].head(100)
flu4_df=flu3_df.groupby(['region']).sum()
del flu4_df['Percent_Positive']
flu4_df.columns=['Number Positive', 'Specimens Tested']
#print(flu4_df)
#print(flu4_df.index)
#ax.bar(flu4_df.index, flu4_df["Number_Positive"], color='r', label="# of Positive Results", width=.9)
#ax.bar(flu4_df.index, flu4_df["Specimens_Tested"], color='b', label="Specimens_Tested", width=.9)
ax = flu4_df.plot.bar(figsize=(15,10), rot=0)
ax.set_xlabel("Region")
ax.set_ylabel("Sum of Count")
ax.legend(loc="best")
#plt.figure(figsize=(200,100))
plt.show()
```
